```python
# Estrutura de decisões

# Escrever um programa que recebe 02 números e identifica o número maior.

# Entrada

numero1 = int(input("digite o primeiro número:"))
numero2 = int(input("digite um número:"))

#Processamento e Saída
if numero1 > numero2:
  print(f"O número {numero1} é maior que o número {numero2}.")
else:
  print(f"O número {numero2} é maior que o número {numero1}.")
```

```python
numero1 = int(input("digite um número:"))
numero2 = int(input("digite um número:"))

#Processamento e Saída
if numero1 > numero2:
  print(f"O número {numero1} é maior que o número {numero2}.")
elif numero1 == numero2:
  print(f"O número {numero1} é igual {numero2}.")
else:
  print(f"O número {numero1} é menor que o número {numero2}.")
```
